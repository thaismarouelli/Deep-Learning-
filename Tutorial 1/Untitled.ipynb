```python
import numpy as np
```

```python
#defining values for input, hidden and output layers
n_input = 2
n_hidden = 3
n_output = 2
```

```python
#initializing values (vector) for the weights and biases

#for input layer to the hidden layer 
np.random.seed(0)
W1 = np.random.randn(n_input, n_hidden)  #weitght
b1 = np.zeros((1, n_hidden))             #bias 

#for hidden layer to the output layer 
W2 = np.random.randn(n_hidden, n_output) #weight
b2 = np.zeros((1, n_output))             #bias 
```

```python
#defining the nonlinear activation fuction: sigmnoid function 
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

#defining the Derivative of Sigmoid
def sigmoid_deriv(x):
    return x * (1 - x)

#defining the loss function as the Mean Squared Error (MSE) between the predicted outputs and the true labels
def mse_loss(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)
```

```python
#setting up data
X = np.array([[0,0],    #input data 
              [0,1],
              [1,0],
              [1,1]])

y = np.array([[0,0],    #target output 
              [1,-1],
              [1,1],
              [2,0]])

#normilizing output into [0, 1] 
y = (y - y.min()) / (y.max() - y.min())
```

```python
#training setup and learning loop

lr = 0.5        #learning rate: how big a step the network takes when updating weights
epochs = 10000  #number of full passes through all training samples.

for epoch in range(epochs):
    z1 = X.dot(W1) + b1     #hidden layer 
    a1 = sigmoid(z1)
    z2 = a1.dot(W2) + b2    #output layer 
    a2 = sigmoid(z2)
    loss = mse_loss(y, a2)  #compute loss 
    d_a2 = (a2 - y)         #Backpropagation
    d_z2 = d_a2 * sigmoid_deriv(a2)
    d_W2 = a1.T.dot(d_z2)
    d_b2 = np.sum(d_z2, axis=0, keepdims=True)
    d_a1 = d_z2.dot(W2.T)
    d_z1 = d_a1 * sigmoid_deriv(a1)
    d_W1 = X.T.dot(d_z1)
    d_b1 = np.sum(d_z1, axis=0, keepdims=True)
    W2 -= lr * d_W2        #Update Parameters
    b2 -= lr * d_b2
    W1 -= lr * d_W1
    b1 -= lr * d_b1
    if epoch % 2000 == 0:   #Monitor progress
        print(f"Epoch {epoch}, Loss: {loss:.4f}")

print("\nPredictions after training:")
print(a2)
```
